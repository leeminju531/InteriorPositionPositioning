## 2. RSSI값은 같은 거리에 위치했을지라도, 측정 세기의 평균과 표준편차가 상이하다. 이에, Normalization에서는 StandardScaler가 더 적합해 보인다.




## 비콘의 신호세기 데이터를 수집하는데 있어서, 비콘 사이의 장애물이 존재하지 않도록 유지시켜 주었다. 그럼에도 불구하고, 원거리의 비콘은 상당한 표준편차가 발생하였는데, 이는 맵위의 비콘이 벽에 반사된 신호세기를 측정하였거나, 각각의 블루투스 신호 세기의 방향이 하늘을 향하고 있어 더 큰 오차를 초래한 것 같다.
